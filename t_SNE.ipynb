```python
#from pandas import read _csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.manifold import TSNE
from bioinfokit.visuz import cluster

```

```python
#Load data
df=pd.read_csv('/content/TSNE_data.csv')
```

```python
df.head() #M-Malignant-cancer tumor present , B Benign - not a cancer tumor. We want to find if there is a pattern in M tumor and B tumor
```

```python
#split out validation dataset
array=df.values
# separate array into input and output components
X=array[:,1:]
Y=array[:,0]
```

```python
df
```

```python
#T-SNE visualization
data_tsne=TSNE(n_components=2).fit_transform(X) # no of components = 2, max you can go for 3
cluster.tsneplot(score=data_tsne)  #vizualize data
```

```python
data_tsne
```

```python
newdf=pd.DataFrame(data_tsne).join(df["diagnosis"])
newdf
```

```python
#get a list of categories
color_class=df['diagnosis'].to_numpy() # Convert dataframe to array. legnedpos=position on upper right corner showing color difference, colorlist-show different color for M and B in diagnosis column
cluster.tsneplot(score=data_tsne,colorlist=color_class,legendpos='upper right',legendanchor=(1.15,1))
plt.show()
#plot will be stored in default directory
```
