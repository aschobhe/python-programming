```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
beml_df=pd.read_csv('BEML.csv')
beml_df.head()
```

```python
glaxo_df=pd.read_csv('GLAXO.csv')
glaxo_df.head()
```

```python
beml_df=beml_df[['Date','Close']]
glaxo_df=glaxo_df[['Date','Close']]
```

```python
beml_df
```

```python
beml_df.dtypes
```

```python
glaxo_df.dtypes
```

```python
#the dataframe have date column,so we can create a datetime index from this column date.it will ensure that rows are 
#sorted by time in ascending order
```

```python
glaxo_df=glaxo_df.set_index(pd.DatetimeIndex(glaxo_df['Date']))
beml_df=beml_df.set_index(pd.DatetimeIndex(beml_df['Date']))
```

```python
glaxo_df
```

```python
beml_df # date column act as index
```

```python
glaxo_df.head(20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
plt.plot(glaxo_df.Close)
plt.xlabel('time')
plt.ylabel('close price')
```

```python
plt.plot(beml_df.Close)
plt.xlabel('time')
plt.ylabel('close')
```

```python
#pct_change=percentage formula
glaxo_df['gain']=glaxo_df.Close.pct_change(periods=1)
beml_df['gain']=beml_df.Close.pct_change(periods=1)
```

```python
glaxo_df
```

```python
beml_df
```

```python
#drop first row since it is  Nan
glaxo_df=glaxo_df.dropna()
beml_df=beml_df.dropna()
```

```python
glaxo_df
```

```python
beml_df
```

```python
#plot the gain
plt.figure (figsize=(8,5))
plt.plot(glaxo_df.gain);
plt.xlabel('time');
plt.ylabel('gain');
```

```python
plt.figure (figsize=(8,5))
plt.plot(beml_df.gain);
plt.xlabel('time');
plt.ylabel('gain');
```

```python
import seaborn as sn
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.suptitle('distribution',fontsize=18)

sn.distplot(glaxo_df.gain,label='GLAXO')

plt.xlabel('gain')
plt.ylabel('density')


plt.subplot(1,2,2)


sn.distplot(beml_df.gain,label='BEML')

plt.xlabel('gain')
plt.ylabel('density')
```

```python
print('mean:',round(glaxo_df.gain.mean(),4))
print('standard deviation:',round(glaxo_df.gain.std(),4))
```

```python
print('mean:',round(beml_df.gain.mean(),4))
print('standard deviation:',round(beml_df.gain.std(),4))
```
